```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df1 = pd.read_csv('transactions.csv')
# df2 = pd.read_csv('transactions1.csv')

# df = pd.concat([df1, df2])
df1
```

```python
df1.Category.unique()

```

```python
# format and clean the stuff

df1.Category.replace(['Clothing' , 'Books', 'Shopping', 'Furnishings', 'Home Improvement', 'Movies & DVDs',
                     'Gifts & Donations','Uncategorized', 'Personal Care', 'Misc Expenses', 'Service & Parts', 
                     'Vacation', 'Electronics & Software','Home', 'Sporting Goods', 'Hobbies', 'Entertainment', 
                     'Home Services', 'Hair', 'Sports', 'Charity'],'Shopping',inplace=True)

df1.Category.replace(['Fast Food', 'Restaurants', 'Food & Dining', 'Coffee Shops', 'Food Delivery'],
                    'Restaurants', inplace=True)

df1.Category.replace(['Gas & Fuel', 'Ride Share', 'Parking', 'Travel', 'Hotel'], 'Transportation', inplace=True)

df1.Category.replace(['Groceries', 'Pet Food & Supplies', 'Laundry'], 'Groceries', inplace=True)

df1.Category.replace(['Pharmacy', 'Utilities', 'Auto Insurance', 'Tuition', 'Mortgage & Rent', 'Home Insurance', 'Internet', 
                     'Gym', 'Television', 'Music'], 'Utilities', inplace=True)

df1.Category.replace(['Transfer', 'Bank Fee', 'Credit Card Payment', 'Investments', 'ATM Fee', 'Cash & ATM', 'Financial', 
                     'Business Services'], 'Finance', inplace=True)

df1.Category.replace(['Alcohol & Bars'], 'Bar', inplace=True)

df1.Category.replace(['Paycheck', 'Income'], 'Income', inplace=True)
```

```python
df1.Category.unique()
```

```python
df = df1.drop(['Original Description', 'Transaction Type', 'Account Name', 'Labels', 'Notes', 'C'], axis=1)
```
